```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    separators=["\n", "\n\n", " ", ""],
    chunk_size=100,
    chunk_overlap=20,
    length_function=len
)
```

```python
chunks = text_splitter.split_text(document[0].page_content)
```
